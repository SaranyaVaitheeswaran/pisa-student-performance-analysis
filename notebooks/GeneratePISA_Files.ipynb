```python
def cols_to_keep_and_match(csvDataFile, requested_cols):

    # 1. Read just the header of the target file (nrows=0 is nearly instant)
    available_cols = pd.read_csv(csvDataFile, nrows=0).columns.tolist()
    
    # 2. Get the requested columns from your list file
    #requested_cols = pd.read_csv('columns_list.csv').columns.tolist()
    
    # 3. Only keep columns that actually exist in the large file
    cols_to_load = [colName for colName in requested_cols if colName in available_cols]

    '''
    # 4. Load ONLY the required columns from the large file This saves significant time and RAM
    #df = pd.read_csv('your_data.csv', usecols=cols_to_load)
    # 5. Create the filtered dataframe
    #filtered_df = df[cols_to_load]
    '''
    #return filtered_df

    return cols_to_load


# Main Program
subjects_cols_to_import = [f'PV{i}{sub}' for sub in ['MATH', 'READ', 'SCIE'] for i in range(1, 11)]

# 1. Load the column names from the first row of your column file
# pandas reads the first row as columns by default
fileHavingColumns = os.path.join("DataSet","PISA_Fields_List.csv")
requested_cols = pd.read_csv(fileHavingColumns).columns.tolist()
print('\nList of Columns\n', requested_cols)


fileToRead_22 = "CY08MSP_STU_QQQ.csv"
fileToRead_18 = "cy07_msu_stu_qqq.csv"

filesToRead = {'PISA_2022': 'CY08MSP_STU_QQQ.csv', 'PISA_2018': 'cy07_msu_stu_qqq.csv'}
filesToSave = {'PISA_2022': 'PISA_2022_WithMeanScore.csv', 'PISA_2018': 'PISA_2018_WithMeanScore.csv'}

for fileCode in filesToRead.keys():
    #get the file name from items
    csvFileNameToRead = os.path.join("DataSet",filesToRead[fileCode])
    csvFileNameToWrite = os.path.join("DataSet",filesToSave[fileCode])
    
    #get the matching columns
    cols_to_import = cols_to_keep_and_match(csvFileNameToRead,requested_cols)
    print("Columns List.. that exists are..\n")
    print(cols_to_import)

    #now read the full data file for only matching columns
    print("Now..Reading Full FIle..based on available columns")
    # Replace 'your_file.csv' with your actual filename
    df_PISA = pd.read_csv(csvFileNameToRead, usecols=cols_to_import)
    df_PISA = df_PISA.dropna(subset=subjects_cols_to_import)
    # 2. Group the Plausible Values (PV1-PV10)
    subjects = {'Math': 'MATH', 'Reading': 'READ', 'Science': 'SCIE'}
    
    for label, sub_code in subjects.items():
        pv_cols = [f'PV{i}{sub_code}' for i in range(1, 11)]
        # Create a student-level average across the 10 PVs
        df_PISA[f'{label}_MeanScore'] = df_PISA[pv_cols].mean(axis=1)

    #Save File
    print("Saving  FIle...")
    df_PISA.to_csv(csvFileNameToWrite, index=False)

    #Print success 
    print("File Saved ..", csvFileNameToWrite)
```
